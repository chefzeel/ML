```python
label=studentdata['race/ethnicity'].value_counts().index
colors=['blue','red','yellow','green','brown']
explode=[0,0,0.1,0,0]
values=studentdata['race/ethnicity'].value_counts().values

plt.figure(figsize=(7,7))
plt.pie(values,explode=explode,labels=label,colors=colors,autopct='%1.1f%%')
plt.title('Race/Ethnicity Analysis',color='black',fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(4,4))
label=['math score','reading score','writing score']
colors=['blue','red','yellow']
explode=[0,0,0.1]
values=[studentdata['math score'].mean(),
        studentdata['reading score'].mean(),
        studentdata['writing score'].mean()
       ]
plt.pie(values,explode=explode,labels=label,colors=colors,autopct='%1.1f%%')
plt.title('Reading Score For Every Race/Ethnicity Mean',color='black',fontsize=10)
plt.axis('equal')
plt.show()
```
